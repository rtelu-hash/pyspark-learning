```python
# WHERE salary > 400000
high_earners = df[df["salary"] > 400000]
print(high_earners)
```

```python
# SELECT name, salary ORDER BY salary DESC
df_sorted = df[["name", "salary"]].sort_values("salary", ascending=False)
print(df_sorted)
```

```python
# SELECT specific columns (like SELECT name, department, city)
df_selected = df[["name", "department", "city"]]
print(df_selected)
```

```python
# ORDER BY salary ASC
df_asc = df.sort_values("salary", ascending=True)
print(df_asc[["name", "salary"]])
```

```python
# SELECT department, COUNT(*) FROM employees GROUP BY department
dept_count = df.groupby("department")["name"].count()
print(dept_count)
```

```python
dept_count = df.groupby("department")["name"].count()
print(dept_count)
```

```python
dept_avg = df.groupby("department")["salary"].mean()
print(dept_avg)
```

```python
dept_summary = df.groupby("department")["salary"].agg(["sum", "mean", "max", "min"])
print(dept_summary)
```

```python
# Save the department summary to a new CSV file
dept_summary.to_csv("department_summary.csv")
print("File saved!")
```

```python
# Save filtered data — only high earners
high_earners = df[df["salary"] > 400000]
high_earners.to_csv("high_earners.csv", index=False)
print("High earners file saved!")
```
