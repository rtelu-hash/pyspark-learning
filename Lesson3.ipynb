````python
import pandas as pd

# Load the messy data
df = pd.read_csv("messy_employees.csv")
print(df)
print("\nShape:", df.shape)
````

````python
# Count missing values in each column (like SELECT COUNT(*) WHERE col IS NULL)
print(df.isnull().sum())
````

````python
# See which rows have ANY missing value
print(df[df.isnull().any(axis=1)])
````

````python
# Fill missing salary with the AVERAGE salary (like COALESCE in SQL)
avg_salary = df["salary"].mean()
df["salary"].fillna(avg_salary, inplace=True)
print("Missing salary filled with average:", avg_salary)
````

````python
# Convert salary column to numeric - force errors to NaN
df["salary"] = pd.to_numeric(df["salary"], errors="coerce")
print(df["salary"])
````

````python
# Fill missing salary with average
avg_salary = df["salary"].mean()
df["salary"].fillna(avg_salary, inplace=True)
print("Missing salary filled with average:", avg_salary)
````

````python
# Fill missing salary with average
avg_salary = df["salary"].mean()
df["salary"] = df["salary"].fillna(avg_salary)
print("Missing salary filled with average:", avg_salary)
````

````python
# Fill missing department and city
df["department"] = df["department"].fillna("Unknown")
df["city"] = df["city"].fillna("Unknown")
````

````python
# Drop rows where name is missing
df = df.dropna(subset=["name"])
print(df)
````

````python
# Fix inconsistent casing
df["department"] = df["department"].str.title()
print(df["department"].unique())
````
